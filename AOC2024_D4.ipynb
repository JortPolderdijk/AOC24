```python
with open('input-d4p1.txt', 'r') as file:
  input = [line.strip() for line in file]
input
```

```python
%%time

search = "XMAS"

count = 0

for y in range(len(input)):
  for x in range(len(input[0])):
    for dy in [-1, 0, 1]:
      for dx in [-1, 0, 1]:
        if dx == 0 and dy == 0:
          continue
        found = True
        for i in range(len(search)):
          lookx, looky = x + i * dx, y + i * dy
          if not (0 <= lookx < len(input[0]) and 0 <= looky < len(input) and input[looky][lookx] == search[i]):
            found = False
            break
        if found:
          count += 1

print("Output:")
print(count)
```
